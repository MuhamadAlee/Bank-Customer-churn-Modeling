```python
import pandas as pd
```

```python
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df
```

```python
df.drop(['RowNumber','CustomerId','Surname'], inplace=True, axis=1)
```

```python
df
```

```python
df.isna().any()
```

```python
df.describe()
```

```python
###### Removing Outlier
```

```python
max_threshold=df['CreditScore'].mean()+3*df['CreditScore'].std()
min_threshold=df['CreditScore'].mean()-3*df['CreditScore'].std()
```

```python
df1=df[(df['CreditScore']>min_threshold) & (df['CreditScore']<max_threshold)]
```

```python
df1
```

```python
max_threshold=df['Age'].mean()+3*df['Age'].std()
min_threshold=df['Age'].mean()-3*df['Age'].std()
```

```python
df2=df1[(df1['Age']>min_threshold) & (df1['Age']<max_threshold)] 
```

```python
df2
```

```python
max_threshold=df['Tenure'].mean()+3*df['Tenure'].std()
min_threshold=df['Tenure'].mean()-3*df['Tenure'].std()
```

```python
df3=df2[(df2['Tenure']>min_threshold) & (df2['Tenure']<max_threshold)] 
```

```python
df3
```

```python
max_threshold=df['Balance'].mean()+3*df['Balance'].std()
min_threshold=df['Balance'].mean()-3*df['Balance'].std()
```

```python
df4=df3[(df3['Balance']>min_threshold) & (df3['Balance']<max_threshold)] 
```

```python
df4
```

```python
df5=pd.get_dummies(df4,columns=['Geography','Gender'],drop_first=True)
```

```python
df5
```

```python
y=df5['Exited']
X=df5.drop(columns=['Exited'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(output_dim=16, init='uniform', activation='relu',input_dim=11))
model.add(Dropout(p=1.0))

#adding the second hidden layer
model.add(Dense(output_dim=16, init='uniform', activation='relu'))
model.add(Dropout(p=1.0))

#adding the output layer
model.add(Dense(output_dim=1, init='uniform', activation='sigmoid'))

```

```python
model.compile(optimizer="Adam", loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=100, nb_epoch=150)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
